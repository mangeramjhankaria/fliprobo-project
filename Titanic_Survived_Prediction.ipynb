# Data Cleaning

```python
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
def input_age(cols):
    Age = cols[0]
    Pclass= cols[1]
    if pd.insull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age'].isnull().sum()
```

```python
train[['Age','Pclass']]
```

```python
train['Age']=train[['Age','Pclass']].apply(input_age,axis =1)
```
